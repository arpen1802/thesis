```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the logged metrics, specify the path
metrics = pd.read_csv('logs/dnn-dur/version_12/metrics.csv')
# Extract the relevant columns
# Pick train_loss_epoch or train_loss depending on how often the loss was logged
train_loss_delta_soc = metrics['train_loss_epoch'].values
train_loss_delta_soc = train_loss_delta_soc[~np.isnan(train_loss_delta_soc)]
val_loss_plugin_duration = metrics['val_loss'].values
val_loss_plugin_duration = val_loss_plugin_duration[~np.isnan(val_loss_plugin_duration)]

epochs = range(1, len(train_loss_delta_soc) + 1)

# Plot the learning curves
plt.figure(figsize=(8, 6))
plt.plot(epochs, train_loss_delta_soc, label='Training Loss', color='blue')
plt.plot(epochs, val_loss_plugin_duration, label='Validation Loss', color='orange')

# plt.plot(epochs, val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Learning Curve for Plugin Duration with DNN')
plt.legend()
plt.grid(True)
# plt.savefig("plots/dnn-dur.png", dpi=300)
plt.show()
```
